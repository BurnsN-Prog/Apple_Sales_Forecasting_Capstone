```python
# Transposing the DataFrame to make columns rows, and rows columns
df_transposed = df.T

new_header = df_transposed.iloc[0]  
df_transposed = df_transposed[1:]   
df_transposed.columns = new_header 

df_transposed.reset_index(inplace=True)
df_transposed.rename(columns={'index': 'Date'}, inplace=True)


df_transposed.head()
```

```python
# Converting numerical strings to floats, removing commas, and reducing values

for col in df_transposed.columns[1:]: 
    df_transposed[col] = df_transposed[col].str.replace(',', '').astype(float)/ 1_000_000


df_transposed.head()
```

```python
# Keeping only the 'Date' and 'TotalRevenue' columns
df = df_transposed[['Date', 'TotalRevenue']]

df.head()
```

```python
df.info()
```

```python
null_values = df.isnull()
null_values.sum()
```

```python
from matplotlib.dates import DateFormatter

# Removing 'ttm'
df = df[df['Date'] != 'ttm']

# Converting the 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Setting the 'Date' column as the index
df.set_index('Date', inplace=True)

# Sorting the DataFrame by the index (Date)
df.sort_index(inplace=True)
```

```python
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['TotalRevenue'])

plt.title('Apple Quarterly Total Revenue (In Millions)')
plt.xlabel('Date')
plt.ylabel('Total Revenue')

plt.gca().xaxis.set_major_formatter(DateFormatter('%Y-%m'))
plt.xticks(rotation=45) 

plt.tight_layout() 
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Augmented Dickey-Fuller test
result = adfuller(df['TotalRevenue'])

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Seasonal decomposition
decomposition = seasonal_decompose(df['TotalRevenue'], model='additive', period=4)


fig, ax = plt.subplots(4, 1, figsize=(12, 8), sharex=True)
decomposition.observed.plot(ax=ax[0], title='Observed')
ax[0].set_ylabel('Revenue')
decomposition.trend.plot(ax=ax[1], title='Trend')
ax[1].set_ylabel('Trend')
decomposition.seasonal.plot(ax=ax[2], title='Seasonality')
ax[2].set_ylabel('Seasonality')
decomposition.resid.plot(ax=ax[3], title='Residuals')
ax[3].set_ylabel('Residuals')
plt.tight_layout()
plt.show()
```

```python
residuals = decomposition.resid.dropna()  

# Calculating the mean and standard deviation of the residuals
residuals_mean = residuals.mean()
residuals_std = residuals.std()

print("Mean of residuals:", residuals_mean)
print("Standard deviation of residuals:", residuals_std)
```

```python
import matplotlib.pyplot as plt

rolling_window = 21

# Calculating the moving average for the last 21 quarters
df['21Q MA'] = df['TotalRevenue'].rolling(window=rolling_window).mean()

# Truncating the data to start from 5 years and 3 months ago, which is 21 quarters
start_period = df.index.max() - pd.DateOffset(years=5, months=3)

df_recent = df.loc[start_period:]
plt.figure(figsize=(12, 6))
plt.plot(df_recent['TotalRevenue'], label='Total Revenue')
plt.plot(df_recent['21Q MA'], label='21-Quarter Moving Average', color='orange')

plt.title('Apple Quarterly Total Revenue & 21-Quarter Moving Average (2019-2024)')
plt.xlabel('Date')
plt.ylabel('Total Revenue (in millions)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
forecast_horizon = 21  

# Splitting the data. Everything other than the last 21 quarters will be used as the training set.
train = df.iloc[:-forecast_horizon]  
validation = df.iloc[-forecast_horizon:] 
```

```python
import pmdarima as pm

# Training the model with the training set
# The m parameter within auto_arima represents the seasonal period (4 for quarterly)
model = pm.auto_arima(train['TotalRevenue'], seasonal=True, m=4, 
                      d=None, D=None, max_p=3, max_q=3, max_P=3, max_Q=3,
                      trace=True, error_action='ignore', suppress_warnings=True)

print(model.summary())
```

```python
import statsmodels.api as sm

# Plotting the residuals to test how well the model captures the patterns within the data
fitted_model = model.arima_res_
residuals = fitted_model.resid

fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# Plot residuals over time
ax[0, 0].plot(residuals)
ax[0, 0].set_title('Residuals Over Time')

# Plot histogram of residuals
ax[0, 1].hist(residuals, bins=25)
ax[0, 1].set_title('Histogram of Residuals')

# Q-Q plot
sm.qqplot(residuals, line='s', ax=ax[1, 0])
ax[1, 0].set_title('Q-Q Plot')

# ACF plot
sm.graphics.tsa.plot_acf(residuals, ax=ax[1, 1])
ax[1, 1].set_title('ACF of Residuals')

plt.tight_layout()
plt.show()
```

```python
# Determining the forecast and convidence values

model.fit(df['TotalRevenue']) 
forecast, conf_int =model.predict(n_periods=len(validation), return_conf_int=True)
future_dates = pd.date_range(start=df.index[-1], periods=forecast_horizon + 1, freq='Q')[1:]


forecast_df = pd.DataFrame({
    'forecast': forecast,
    'lower_conf': conf_int[:, 0],
    'upper_conf': conf_int[:, 1]
}, index=future_dates)


print(forecast_df)
```

```python
# Plotting the forecast with the historical data
plt.figure(figsize=(12, 6))
plt.plot(df['TotalRevenue'], label='Historical Revenue', color='blue')
plt.plot(forecast_df.index, forecast_df['forecast'], label='Forecasted Revenue', color='orange')
plt.fill_between(forecast_df.index,
                 forecast_df['lower_conf'],
                 forecast_df['upper_conf'],
                 color='pink', alpha=0.3)
plt.title('Forecasted Revenue with Confidence Intervals')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
# Taking a closer look at forecast values
plt.figure(figsize=(10, 6))
plt.plot(forecast_df.index, forecast_df['forecast'], label='Forecasted Revenue', color='orange')
plt.fill_between(forecast_df.index,
                 forecast_df['lower_conf'],
                 forecast_df['upper_conf'],
                 color='pink', alpha=0.3)
plt.title('Forecasted Revenue (20 Quarters)')
plt.xlabel('Date')
plt.ylabel('Revenue (in millions)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

actuals = validation['TotalRevenue'].values 

# Calculating the error metrics
mae = mean_absolute_error(actuals, forecast)
mse = mean_squared_error(actuals, forecast)
rmse = np.sqrt(mse)

# To calculate MAPE, you'll need to handle it separately since it's not part of sklearn
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return mape

mape = mean_absolute_percentage_error(actuals, forecast)

# Printing the error metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")
```
